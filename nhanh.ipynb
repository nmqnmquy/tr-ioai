```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import time

# Load model YOLOv8 đã huấn luyện
model = YOLO("run/best.pt")

def detect_mango(image):
    if image is None:
        return None, "⚠️ Vui lòng tải lên một hình ảnh"

    start_time = time.time()

    # Dự đoán
    results = model.predict(image, conf=0.5)

    # Thời gian xử lý
    processing_time = round(time.time() - start_time, 2)

    # Lấy ảnh kết quả từ YOLO (BGR -> cần chuyển sang RGB)
    img_bgr = results[0].plot()

    # Đảm bảo dữ liệu là uint8
    if img_bgr.dtype != np.uint8:
        img_bgr = (img_bgr * 255).astype(np.uint8)

    # Chuyển BGR -> RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Chuyển sang PIL để trả về Gradio
    result_image = Image.fromarray(img_rgb)

    # Thông tin chi tiết
    detections = results[0].boxes
    info_text = f"🕐 **Thời gian xử lý**: {processing_time}s\n\n"

    if detections is not None and len(detections) > 0:
        info_text += f"🔍 **Số lượng xoài phát hiện**: {len(detections)}\n\n"
        info_text += "📊 **Chi tiết từng trái**:\n"

        for i, detection in enumerate(detections):
            conf = float(detection.conf)
            cls = int(detection.cls)
            class_name = model.names[cls] if cls < len(model.names) else "Unknown"
            info_text += f"   • Trái {i+1}: **{class_name}** (Độ tin cậy: {conf:.1%})\n"
    else:
        info_text += "❌ **Không phát hiện được xoài nào trong ảnh**\n"
        info_text += "💡 **Gợi ý**: Thử với ảnh có xoài rõ ràng hơn"

    return result_image, info_text



def clear_all():
    """Reset tất cả"""
    return None, None, ""

# CSS tùy chỉnh cho giao diện
custom_css = """
/* Tùy chỉnh toàn bộ giao diện */
.gradio-container {
    background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
}

/* Header chính */
.main-header {
    text-align: center;
    background: rgba(255, 255, 255, 0.1);
    backdrop-filter: blur(10px);
    border-radius: 20px;
    padding: 30px;
    margin-bottom: 30px;
    border: 1px solid rgba(255, 255, 255, 0.2);
    box-shadow: 0 8px 32px rgba(0, 0, 0, 0.1);
}

.main-header h1 {
    color: white;
    font-size: 2.5em;
    margin-bottom: 10px;
    text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.3);
}

.main-header p {
    color: rgba(255, 255, 255, 0.9);
    font-size: 1.2em;
    margin: 5px 0;
}

/* Các component chính */
.input-section, .output-section {
    background: rgba(255, 255, 255, 0.95);
    border-radius: 15px;
    padding: 25px;
    margin: 15px 0;
    box-shadow: 0 10px 25px rgba(0, 0, 0, 0.1);
    border: 1px solid rgba(255, 255, 255, 0.3);
}

/* Buttons */
.custom-button {
    background: linear-gradient(45deg, #FF6B6B, #4ECDC4);
    border: none;
    border-radius: 25px;
    color: white;
    padding: 12px 30px;
    font-size: 16px;
    font-weight: bold;
    cursor: pointer;
    transition: all 0.3s ease;
    box-shadow: 0 4px 15px rgba(0, 0, 0, 0.2);
}

.custom-button:hover {
    transform: translateY(-2px);
    box-shadow: 0 6px 20px rgba(0, 0, 0, 0.3);
}

/* Image components */
.image-upload {
    border: 3px dashed #4ECDC4;
    border-radius: 15px;
    background: rgba(78, 205, 196, 0.1);
    transition: all 0.3s ease;
}

.image-upload:hover {
    border-color: #FF6B6B;
    background: rgba(255, 107, 107, 0.1);
    transform: scale(1.02);
}


/* Info panel */
.info-panel {
    background: #ffffff; /* đổi thành màu trắng */
    color: #333333;       /* text màu tối để dễ đọc */
    border-radius: 15px;
    padding: 20px;
    margin-top: 15px;
    box-shadow: 0 5px 15px rgba(0, 0, 0, 0.1); /* làm nhẹ bóng đổ */
}


/* Responsive */
@media (max-width: 768px) {
    .main-header h1 {
        font-size: 2em;
    }

    .input-section, .output-section {
        padding: 15px;
        margin: 10px 0;
    }
}
"""

# Tạo giao diện với Blocks
with gr.Blocks(
    css=custom_css,
    title="🥭 AI Mango Classifier",
    theme=gr.themes.Soft(
        primary_hue="emerald",
        secondary_hue="blue",
        neutral_hue="slate",
        font=gr.themes.GoogleFont("Inter")
    )
) as demo:

    # Header
    with gr.Row():
        gr.HTML("""
        <div class="main-header">
            <h1>🥭 Hệ Thống Phân Loại Xoài Thông Minh</h1>
            <p>🤖 Được phát triển bằng mô hình YOLOv8 và AI</p>
            <p>🎯 Phân loại trạng thái chín của trái xoài với độ chính xác cao</p>
        </div>
        """)

    # Main content
    with gr.Row(equal_height=True):
        # Cột trái - Input
        with gr.Column(scale=1, elem_classes="input-section"):
            gr.HTML("<h3 style='text-align: center; color: #333; margin-bottom: 20px;'>📤 Tải Ảnh Lên</h3>")

            input_image = gr.Image(
                label="Chọn ảnh trái xoài",
                type="pil",
                elem_classes="image-upload",
                height=300
            )

            with gr.Row():
                predict_btn = gr.Button(
                    "🔍 Phân Tích Xoài",
                    variant="primary",
                    elem_classes="custom-button",
                    scale=2
                )
                clear_btn = gr.Button(
                    "🗑️ Xóa",
                    variant="secondary",
                    scale=1
                )

            # Thông tin hướng dẫn
            gr.HTML("""
            <div style="background: #f8f9fa; padding: 15px; border-radius: 10px; margin-top: 15px;">
                <h4 style="color: #495057; margin-bottom: 10px;">💡 Hướng Dẫn Sử Dụng:</h4>
                <ul style="color: #6c757d; margin: 0; padding-left: 20px;">
                    <li>Chọn ảnh trái xoài rõ nét</li>
                    <li>Ảnh nên có ánh sáng tốt</li>
                    <li>Xoài nên nằm ở trung tâm ảnh</li>
                    <li>Hỗ trợ các định dạng: JPG, PNG, JPEG</li>
                </ul>
            </div>
            """)

        # Cột phải - Output
        with gr.Column(scale=1, elem_classes="output-section"):
            gr.HTML("<h3 style='text-align: center; color: #fff; margin-bottom: 20px;'>📊 Kết Quả Phân Tích</h3>")

            output_image = gr.Image(
                label="Kết quả phân loại",
                type="pil",  # Quan trọng để không bị lệch màu
                height=300
            )

            # Thông tin chi tiết
            info_output = gr.Markdown(
                value="👋 Tải ảnh lên và nhấn **Phân Tích Xoài** để bắt đầu!",
                elem_classes="info-panel"
            )

    # Footer với thông tin bổ sung
    with gr.Row():
        gr.HTML("""
        <div style="text-align: center; background: rgba(255,255,255,0.1); padding: 20px; border-radius: 15px; margin-top: 20px;">
            <p style="color: white; margin: 5px 0;">
                ⚡ <strong>Mô hình</strong>: YOLOv8 |
                🎯 <strong>Độ chính xác</strong>: 95%+ |
                🚀 <strong>Tốc độ</strong>: <1 giây
            </p>
            <p style="color: rgba(255,255,255,0.8); margin: 5px 0; font-size: 0.9em;">
                📡 Được huấn luyện trên bộ dữ liệu Roboflow với hàng nghìn ảnh xoài
            </p>
        </div>
        """)

    # Event handlers
    predict_btn.click(
        fn=detect_mango,
        inputs=[input_image],
        outputs=[output_image, info_output],
        show_progress=True
    )

    clear_btn.click(
        fn=lambda: (None, None, "👋 Tải ảnh lên và nhấn **Phân Tích Xoài** để bắt đầu!"),
        outputs=[input_image, output_image, info_output]
    )

    # Auto-predict khi upload ảnh
    input_image.change(
        fn=detect_mango,
        inputs=[input_image],
        outputs=[output_image, info_output],
        show_progress=True
    )

# Launch với cấu hình tối ưu
if __name__ == "__main__":
    demo.launch(
        share=True,
        show_error=True,
        quiet=False,
        inbrowser=True,  # Tự động mở browser
        # Bỏ server_port để Gradio tự động tìm port trống
    )
```

```python
import gradio as gr
from ultralytics import YOLO

# Tải mô hình đã huấn luyện (thay đường dẫn nếu khác)
model = YOLO("runs/best.pt")

# Hàm dự đoán
def detect_mango(image):
    results = model.predict(image, conf=0.5)
    # Trả về ảnh có khung vẽ
    return results[0].plot()

# Tạo giao diện Gradio
demo = gr.Interface(
    fn=detect_mango,
    inputs=gr.Image(type="pil", label="Tải ảnh trái xoài"),
    outputs=gr.Image(type="numpy", label="Kết quả phân loại"),
    title="Phân loại trái xoài bằng AI",
    description="Upload ảnh trái xoài để mô hình YOLOv8 phân loại"
)

demo.launch(share=True)  # share=True để tạo link demo có thể chia sẻ
```
