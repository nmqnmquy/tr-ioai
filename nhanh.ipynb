```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import time

# Load model YOLOv8 Ä‘Ã£ huáº¥n luyá»‡n
model = YOLO("run/best.pt")

def detect_mango(image):
    if image is None:
        return None, "âš ï¸ Vui lÃ²ng táº£i lÃªn má»™t hÃ¬nh áº£nh"

    start_time = time.time()

    # Dá»± Ä‘oÃ¡n
    results = model.predict(image, conf=0.5)

    # Thá»i gian xá»­ lÃ½
    processing_time = round(time.time() - start_time, 2)

    # Láº¥y áº£nh káº¿t quáº£ tá»« YOLO (BGR -> cáº§n chuyá»ƒn sang RGB)
    img_bgr = results[0].plot()

    # Äáº£m báº£o dá»¯ liá»‡u lÃ  uint8
    if img_bgr.dtype != np.uint8:
        img_bgr = (img_bgr * 255).astype(np.uint8)

    # Chuyá»ƒn BGR -> RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Chuyá»ƒn sang PIL Ä‘á»ƒ tráº£ vá» Gradio
    result_image = Image.fromarray(img_rgb)

    # ThÃ´ng tin chi tiáº¿t
    detections = results[0].boxes
    info_text = f"ğŸ• **Thá»i gian xá»­ lÃ½**: {processing_time}s\n\n"

    if detections is not None and len(detections) > 0:
        info_text += f"ğŸ” **Sá»‘ lÆ°á»£ng xoÃ i phÃ¡t hiá»‡n**: {len(detections)}\n\n"
        info_text += "ğŸ“Š **Chi tiáº¿t tá»«ng trÃ¡i**:\n"

        for i, detection in enumerate(detections):
            conf = float(detection.conf)
            cls = int(detection.cls)
            class_name = model.names[cls] if cls < len(model.names) else "Unknown"
            info_text += f"   â€¢ TrÃ¡i {i+1}: **{class_name}** (Äá»™ tin cáº­y: {conf:.1%})\n"
    else:
        info_text += "âŒ **KhÃ´ng phÃ¡t hiá»‡n Ä‘Æ°á»£c xoÃ i nÃ o trong áº£nh**\n"
        info_text += "ğŸ’¡ **Gá»£i Ã½**: Thá»­ vá»›i áº£nh cÃ³ xoÃ i rÃµ rÃ ng hÆ¡n"

    return result_image, info_text



def clear_all():
    """Reset táº¥t cáº£"""
    return None, None, ""

# CSS tÃ¹y chá»‰nh cho giao diá»‡n
custom_css = """
/* TÃ¹y chá»‰nh toÃ n bá»™ giao diá»‡n */
.gradio-container {
    background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
}

/* Header chÃ­nh */
.main-header {
    text-align: center;
    background: rgba(255, 255, 255, 0.1);
    backdrop-filter: blur(10px);
    border-radius: 20px;
    padding: 30px;
    margin-bottom: 30px;
    border: 1px solid rgba(255, 255, 255, 0.2);
    box-shadow: 0 8px 32px rgba(0, 0, 0, 0.1);
}

.main-header h1 {
    color: white;
    font-size: 2.5em;
    margin-bottom: 10px;
    text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.3);
}

.main-header p {
    color: rgba(255, 255, 255, 0.9);
    font-size: 1.2em;
    margin: 5px 0;
}

/* CÃ¡c component chÃ­nh */
.input-section, .output-section {
    background: rgba(255, 255, 255, 0.95);
    border-radius: 15px;
    padding: 25px;
    margin: 15px 0;
    box-shadow: 0 10px 25px rgba(0, 0, 0, 0.1);
    border: 1px solid rgba(255, 255, 255, 0.3);
}

/* Buttons */
.custom-button {
    background: linear-gradient(45deg, #FF6B6B, #4ECDC4);
    border: none;
    border-radius: 25px;
    color: white;
    padding: 12px 30px;
    font-size: 16px;
    font-weight: bold;
    cursor: pointer;
    transition: all 0.3s ease;
    box-shadow: 0 4px 15px rgba(0, 0, 0, 0.2);
}

.custom-button:hover {
    transform: translateY(-2px);
    box-shadow: 0 6px 20px rgba(0, 0, 0, 0.3);
}

/* Image components */
.image-upload {
    border: 3px dashed #4ECDC4;
    border-radius: 15px;
    background: rgba(78, 205, 196, 0.1);
    transition: all 0.3s ease;
}

.image-upload:hover {
    border-color: #FF6B6B;
    background: rgba(255, 107, 107, 0.1);
    transform: scale(1.02);
}


/* Info panel */
.info-panel {
    background: #ffffff; /* Ä‘á»•i thÃ nh mÃ u tráº¯ng */
    color: #333333;       /* text mÃ u tá»‘i Ä‘á»ƒ dá»… Ä‘á»c */
    border-radius: 15px;
    padding: 20px;
    margin-top: 15px;
    box-shadow: 0 5px 15px rgba(0, 0, 0, 0.1); /* lÃ m nháº¹ bÃ³ng Ä‘á»• */
}


/* Responsive */
@media (max-width: 768px) {
    .main-header h1 {
        font-size: 2em;
    }

    .input-section, .output-section {
        padding: 15px;
        margin: 10px 0;
    }
}
"""

# Táº¡o giao diá»‡n vá»›i Blocks
with gr.Blocks(
    css=custom_css,
    title="ğŸ¥­ AI Mango Classifier",
    theme=gr.themes.Soft(
        primary_hue="emerald",
        secondary_hue="blue",
        neutral_hue="slate",
        font=gr.themes.GoogleFont("Inter")
    )
) as demo:

    # Header
    with gr.Row():
        gr.HTML("""
        <div class="main-header">
            <h1>ğŸ¥­ Há»‡ Thá»‘ng PhÃ¢n Loáº¡i XoÃ i ThÃ´ng Minh</h1>
            <p>ğŸ¤– ÄÆ°á»£c phÃ¡t triá»ƒn báº±ng mÃ´ hÃ¬nh YOLOv8 vÃ  AI</p>
            <p>ğŸ¯ PhÃ¢n loáº¡i tráº¡ng thÃ¡i chÃ­n cá»§a trÃ¡i xoÃ i vá»›i Ä‘á»™ chÃ­nh xÃ¡c cao</p>
        </div>
        """)

    # Main content
    with gr.Row(equal_height=True):
        # Cá»™t trÃ¡i - Input
        with gr.Column(scale=1, elem_classes="input-section"):
            gr.HTML("<h3 style='text-align: center; color: #333; margin-bottom: 20px;'>ğŸ“¤ Táº£i áº¢nh LÃªn</h3>")

            input_image = gr.Image(
                label="Chá»n áº£nh trÃ¡i xoÃ i",
                type="pil",
                elem_classes="image-upload",
                height=300
            )

            with gr.Row():
                predict_btn = gr.Button(
                    "ğŸ” PhÃ¢n TÃ­ch XoÃ i",
                    variant="primary",
                    elem_classes="custom-button",
                    scale=2
                )
                clear_btn = gr.Button(
                    "ğŸ—‘ï¸ XÃ³a",
                    variant="secondary",
                    scale=1
                )

            # ThÃ´ng tin hÆ°á»›ng dáº«n
            gr.HTML("""
            <div style="background: #f8f9fa; padding: 15px; border-radius: 10px; margin-top: 15px;">
                <h4 style="color: #495057; margin-bottom: 10px;">ğŸ’¡ HÆ°á»›ng Dáº«n Sá»­ Dá»¥ng:</h4>
                <ul style="color: #6c757d; margin: 0; padding-left: 20px;">
                    <li>Chá»n áº£nh trÃ¡i xoÃ i rÃµ nÃ©t</li>
                    <li>áº¢nh nÃªn cÃ³ Ã¡nh sÃ¡ng tá»‘t</li>
                    <li>XoÃ i nÃªn náº±m á»Ÿ trung tÃ¢m áº£nh</li>
                    <li>Há»— trá»£ cÃ¡c Ä‘á»‹nh dáº¡ng: JPG, PNG, JPEG</li>
                </ul>
            </div>
            """)

        # Cá»™t pháº£i - Output
        with gr.Column(scale=1, elem_classes="output-section"):
            gr.HTML("<h3 style='text-align: center; color: #fff; margin-bottom: 20px;'>ğŸ“Š Káº¿t Quáº£ PhÃ¢n TÃ­ch</h3>")

            output_image = gr.Image(
                label="Káº¿t quáº£ phÃ¢n loáº¡i",
                type="pil",  # Quan trá»ng Ä‘á»ƒ khÃ´ng bá»‹ lá»‡ch mÃ u
                height=300
            )

            # ThÃ´ng tin chi tiáº¿t
            info_output = gr.Markdown(
                value="ğŸ‘‹ Táº£i áº£nh lÃªn vÃ  nháº¥n **PhÃ¢n TÃ­ch XoÃ i** Ä‘á»ƒ báº¯t Ä‘áº§u!",
                elem_classes="info-panel"
            )

    # Footer vá»›i thÃ´ng tin bá»• sung
    with gr.Row():
        gr.HTML("""
        <div style="text-align: center; background: rgba(255,255,255,0.1); padding: 20px; border-radius: 15px; margin-top: 20px;">
            <p style="color: white; margin: 5px 0;">
                âš¡ <strong>MÃ´ hÃ¬nh</strong>: YOLOv8 |
                ğŸ¯ <strong>Äá»™ chÃ­nh xÃ¡c</strong>: 95%+ |
                ğŸš€ <strong>Tá»‘c Ä‘á»™</strong>: <1 giÃ¢y
            </p>
            <p style="color: rgba(255,255,255,0.8); margin: 5px 0; font-size: 0.9em;">
                ğŸ“¡ ÄÆ°á»£c huáº¥n luyá»‡n trÃªn bá»™ dá»¯ liá»‡u Roboflow vá»›i hÃ ng nghÃ¬n áº£nh xoÃ i
            </p>
        </div>
        """)

    # Event handlers
    predict_btn.click(
        fn=detect_mango,
        inputs=[input_image],
        outputs=[output_image, info_output],
        show_progress=True
    )

    clear_btn.click(
        fn=lambda: (None, None, "ğŸ‘‹ Táº£i áº£nh lÃªn vÃ  nháº¥n **PhÃ¢n TÃ­ch XoÃ i** Ä‘á»ƒ báº¯t Ä‘áº§u!"),
        outputs=[input_image, output_image, info_output]
    )

    # Auto-predict khi upload áº£nh
    input_image.change(
        fn=detect_mango,
        inputs=[input_image],
        outputs=[output_image, info_output],
        show_progress=True
    )

# Launch vá»›i cáº¥u hÃ¬nh tá»‘i Æ°u
if __name__ == "__main__":
    demo.launch(
        share=True,
        show_error=True,
        quiet=False,
        inbrowser=True,  # Tá»± Ä‘á»™ng má»Ÿ browser
        # Bá» server_port Ä‘á»ƒ Gradio tá»± Ä‘á»™ng tÃ¬m port trá»‘ng
    )
```

```python
import gradio as gr
from ultralytics import YOLO

# Táº£i mÃ´ hÃ¬nh Ä‘Ã£ huáº¥n luyá»‡n (thay Ä‘Æ°á»ng dáº«n náº¿u khÃ¡c)
model = YOLO("runs/best.pt")

# HÃ m dá»± Ä‘oÃ¡n
def detect_mango(image):
    results = model.predict(image, conf=0.5)
    # Tráº£ vá» áº£nh cÃ³ khung váº½
    return results[0].plot()

# Táº¡o giao diá»‡n Gradio
demo = gr.Interface(
    fn=detect_mango,
    inputs=gr.Image(type="pil", label="Táº£i áº£nh trÃ¡i xoÃ i"),
    outputs=gr.Image(type="numpy", label="Káº¿t quáº£ phÃ¢n loáº¡i"),
    title="PhÃ¢n loáº¡i trÃ¡i xoÃ i báº±ng AI",
    description="Upload áº£nh trÃ¡i xoÃ i Ä‘á»ƒ mÃ´ hÃ¬nh YOLOv8 phÃ¢n loáº¡i"
)

demo.launch(share=True)  # share=True Ä‘á»ƒ táº¡o link demo cÃ³ thá»ƒ chia sáº»
```
